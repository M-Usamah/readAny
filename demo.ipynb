```python
from transformers import BarkModel

model = BarkModel.from_pretrained("suno/bark-small")
```

```python
import torch

device = "cuda:0" if torch.cuda.is_available() else "cpu"
model = model.to(device)
```

```python
from transformers import AutoProcessor

processor = AutoProcessor.from_pretrained("suno/bark")
```

```python
# prepare the inputs
text_prompt = "We have claimed that AI is exciting, but we have not said what it is. In Figure 1.1 we see eight definitions of AI, laid out along two dimensions."
```

```python
inputs = processor(text_prompt, voice_preset="v2/en_speaker_9")

```

```python
# # generate speech
speech_output = model.generate(**inputs.to(device))
```

```python
from IPython.display import Audio

sampling_rate = model.generation_config.sample_rate
Audio(speech_output[0].cpu().numpy(), rate=sampling_rate)
```

```python
speech_output = model.generate(**inputs, num_beams = 4, temperature = 0.5, semantic_temperature = 0.8)

Audio(speech_output[0].cpu().numpy(), rate=sampling_rate)
```
